```python
# Final Class Conversion
class FSMatch:
    __shot_left = None
    __caster1 = 0
    __caster2 = 0
    __caster3 = 0 
    __teamATotal = 0
    __teamBTotal = 0

    def __init__(self, shotLeft):
        self.__shot_left = shotLeft


    def __truncateAll(self):
        file2 = open("Caster1.txt","w+")
        file2.truncate()
        file2.close()

        file2 = open("Caster2.txt","w+")
        file2.truncate()
        file2.close()

        file2 = open("Caster3.txt","w+")
        file2.truncate()
        file2.close()

        file2 = open("Total.txt","w+")
        file2.truncate()
        file2.close()

    def __enterScore(self,filePath, caster):
        file2 = open(filePath,"w+")
        file2.truncate()
        caster = int(input(f"Enter Caster {filePath[6]} Score: "))
        file2.write(str(caster))
        file2.close()
        return caster
    
    def teamScoreRefresh(self,a,b):
        file2 = open("TeamAScore.txt","w+")
        file2.truncate()
        file2.write(str(a))
        file2.close()
        self.__teamATotal = a
        self.__teamBTotal = b

        file2 = open("TeamBScore.txt","w+")
        file2.truncate()
        file2.write(str(b))
        file2.close()

    def shots_remaining(self):
        file2 = open("shotLeft.txt","w+")
        file2.truncate()
        file2.write(str(self.__shot_left))
        if(self.__shot_left == "Over"):
            self.__shot_left=11
        self.__shot_left-=1
        file2.close()
        if(self.__shot_left==0):
            self.__shot_left="Over"

    def casterscore(self,isTeamA):
        self.__truncateAll()

        self.__caster1 = self.__enterScore("Caster1.txt", self.__caster1)
        self.__caster2 = self.__enterScore("Caster2.txt", self.__caster2)
        self.__caster3 = self.__enterScore("Caster3.txt", self.__caster3)

        
        if(isTeamA):
            totalA = self.__caster1 + self.__caster2 + self.__caster3

            file2 = open("Total.txt","w+")
            file2.write(str(totalA))
            file2.close()

            if(self.__teamATotal < totalA):
                self.__teamATotal = totalA
                file2 = open("TeamAScore.txt","w+")
                file2.write(str(totalA))
                file2.close()

        if(isTeamA == False):
            totalB = self.__caster1 + self.__caster2 + self.__caster3

            file2 = open("Total.txt","w+")
            file2.write(str(totalB))
            file2.close()

            if(self.__teamBTotal < totalB):
                self.__teamBTotal = totalB
                file2 = open("TeamBScore.txt","w+")
                file2.write(str(totalB))
                file2.close()

    def castor_reset(self):
        self.__caster1 = 0
        self.__caster2 = 0
        self.__caster3 = 0
        self.__truncateAll()


    def display_status(self):
        print(self.__shot_left,self.__teamATotal,self.__teamBTotal,self.__caster1,self.__caster2,self.__caster3)
```

```python
m1 = FSMatch(10)
```

```python
while(True):
    '''
    print("Select Option:")
    print("1. Shots Remaining")
    print("2. Blue Score")
    print("3. Orange Score")
    print("4. Caster Reset")
    print("5. Team Score Reset")
    print("6. Display Status")
    print("0. Exit")
    
    '''
    


    option = input()

    if(option == "0"):
        break
    elif(option == "1"):
        m1.shots_remaining()
    elif(option == "2"):
        m1.casterscore(True)
    elif(option == "3"):
        m1.casterscore(False)
    elif(option == "4"):
        m1.castor_reset()
    elif(option == "5"):
        m1.teamScoreRefresh(0,0)
    elif(option == "6"):
        m1.display_status()
    
```

```python
# Display Status
m1.display_status()
```
